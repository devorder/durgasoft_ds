# How to iterate over elements of nd-array.
1. By using python's loop.
2. By using nditer() function.
3. By using ndenumerate() function.

```python
# Iterate using python loop
# 1d-array
a = np.arange(10, 101, 10)
for i in a:
    print(i)
```

```python
# 2d-array
a = np.arange(10, 91, 10).reshape((3, 3))
for i in a:
    for j in i:
        print(j)
```

```python
# 3d-array
a = np.arange(10, 81, 10).reshape((2,2,2))
for i in a:
    for j in i:
        for k in j:
            print(k)
```

### python's for loop approach is not very effective as we need to use a for loop for each level of array.

# nditer

```python
# 1d-array
a = np.arange(10, 101, 10)
for item in np.nditer(a):
    print(item)
```

```python
# 2d-array
a = np.arange(10, 91, 10).reshape((3, 3))
for item in np.nditer(a):
    print(item)
```

```python
# 3d-array
a = np.arange(10, 81, 10).reshape((2, 2, 2))
for item in np.nditer(a):
    print(item)
```

```python
# Iteratre elements of sliced array
a = np.arange(10, 91, 10).reshape((3, 3))
for item in np.nditer(a[:, :2]):
    print(item)
```

```python
# Chaging data type of elements
a = np.arange(10, 91, 10, dtype='int8').reshape((3, 3))
for item in np.nditer(a, op_dtypes=float, flags=('buffered',)):
    print(item)
for item in np.nditer(a, op_dtypes=str, flags=('buffered',)):
    print(item)
```
